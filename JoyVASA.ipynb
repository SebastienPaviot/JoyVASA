```python
import subprocess
import os

def add_text_to_video(input_video: str, output_video: str, text: str = "François Bayrou"):
    """
    Ajoute un texte centré à droite sur une vidéo.

    Args:
        input_video (str): Chemin vers la vidéo d'entrée.
        output_video (str): Chemin où la vidéo générée sera sauvegardée.
        text (str): Texte à afficher (par défaut : "François Bayrou").
    """
    # Vérifier si le fichier d'entrée existe
    if not os.path.isfile(input_video):
        raise FileNotFoundError(f"La vidéo d'entrée '{input_video}' est introuvable.")

    # Commande FFmpeg pour ajouter du texte
    cmd = [
        "ffmpeg",
        "-i", input_video,  # Fichier d'entrée
        "-vf", f"drawtext=text='{text}':fontcolor=white:fontsize=48:x=w-tw-10:y=(h-text_h)/2",  # Texte centré à droite
        "-codec:a", "copy",  # Ne pas réencoder l'audio
        output_video  # Fichier de sortie
    ]

    # Exécuter la commande
    subprocess.run(cmd, check=True)

# Exemple d'utilisation
add_text_to_video("animations/emmanuel.mp4", "output.mp4","Emmanuel Macron")
```
